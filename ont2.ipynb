<h1>Kolmen vastaajamäärältään suurimman koulutusohjelman tarkastelu</h1>

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tilastoapu import *
%matplotlib inline
```

<h2>Hyödyllisyys tunnuslukuina koulutusohjelman mukaan</h2>

```python
dfx = df.groupby('Pasila|Malmi')[hyödyllinen].describe().T
dfx.columns = toimipiste
dfx
```

<h2>Hyödyllisyys tunnuslukuina opinnäytetyön muodon mukaan</h2>

```python
dfx = df.groupby('Muoto')[hyödyllinen].describe().T
dfx.columns = muoto
dfx
```

<h2>Hyödyllisyyden keskiarvot koulutusohjelman ja opinnäytetyön muodon mukaan</h2>

```python
dfx = df.groupby(['Pasila|Malmi', 'Muoto'])[hyödyllinen].mean().round(2)
dfx.index = dfx.index.set_levels(toimipiste, level=0)
dfx.index = dfx.index.set_levels(muoto, level=1)
sns.heatmap(dfx, annot=True)
```

<h2>Hyödyllisyyden keskiarvot sukupuolen mukaan</h2>

```python
dfx = df.groupby(['Sukupuoli'])[hyödyllinen].mean().round(2)
dfx.index = sukupuoli
sns.heatmap(dfx, annot = True)
```

<h2>Ontprosessi tunnuslukuina opinnäytetyön muodon mukaan</h2>

<h2>Opinnäytetyön hyödyllisyys korreloi merkitsevästi kaikkien ontprosessin tekijöiden kanssa</h2>

```python
from scipy.stats import pearsonr
for var in df[['Työprosesseja', 'Asemia', 'Asiantuntijuutta', 'Ratkaisuja']]:
    df_dropna = df.dropna(subset=[var, 'Opinnäytetyö'])
    print(var, pearsonr(df_dropna['Opinnäytetyö'], df_dropna[var]))
    
# Tulosteessa vierekkäin korrelaatiokerroin ja p-arvo
```
