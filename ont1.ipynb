<h2>Hyödyllisyyden keskiarvot toimipisteen ja opinnäytetyön muodon mukaan</h2>

```python
dfx = df.groupby(['Toimipiste', 'Muoto'])[hyödyllinen].mean().round(2)
dfx.index = dfx.index.set_levels(toimipiste, level=0)
dfx.index = dfx.index.set_levels(muoto, level=1)
sns.heatmap(dfx, annot=True)
```

<h2>Opinnäytetyön hyödyllisyys korreloi merkitsevästi kaikkien ontprosessin tekijöiden kanssa</h2>

```python
from scipy.stats import pearsonr
for var in df[['Työprosesseja', 'Asemia', 'Asiantuntijuutta', 'Ratkaisuja']]:
    df_dropna = df.dropna(subset=[var, 'Opinnäytetyö'])
    print(var, pearsonr(df_dropna['Opinnäytetyö'], df_dropna[var]))
    
# Tulosteessa vierekkäin korrelaatiokerroin ja p-arvo
```
